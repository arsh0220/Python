```python
#WAP TO PRINT WHETHER THE NUMBER IS PRIME OR NOT
n=int(input("Enter the value:"))
d=0
for i in range(1,n):
  if(n%i==0):
    d=d+1
  if(d==2):
    break
if(d==2):
  print("Number is not prime")
else:
  print("Number is prime")
```

```python
#WAP TO PRINT MAX OF THE DIGITS OF NUMBER
x=int(input("Enter the number:"))
max=x%10
while x!=0:
  a=x%10
  x=x//10
  if(a>max):
    max=a
print(max)
```

```python
#WAP TO COUNT THE MAX NUMBER
x=int(input("Enter the number:"))
count=0
max=x%10
while x!=0:
  a=x%10
  x=x//10
  if (max==a):
    count = count+1
  if(a>max):
    max=a
print(max)
print(count)
```
